```python
import requests
import pandas as pd
import numpy as np
import csv
from bs4 import BeautifulSoup
from IPython.display import display
from IPython.display import Image
import re
page = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
page

soup = BeautifulSoup(page.text, 'html.parser')

#print(soup.prettify)

gdp_table=soup.find("table", attrs={"class": "wikitable sortable"})
# gdp_table_data=gdp_table.tbody.find_all("tr")
#gdp_table_data

#print(gdp_table)

gdp_table = soup.find('table')
gdp_table_rows = gdp_table.find_all('tr')
gdp_table

Tabel = []
for tr in gdp_table_rows:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        Tabel.append(row)
        #print(Tabel)

PCode = pd.DataFrame(Tabel, columns=["Postcode","Borough","Neighbourhood"])
PCode.head(10)                            
#PCode.describe()

#PCode.shape

PCode1=PCode.loc[PCode['Borough']!='Not assigned']
PCode1.head(5)

#PCode1.shape
PCode1 = PCode1.groupby(['Postcode', 'Borough']).agg(', '.join)
PCode1 = PCode1.reset_index()
#PCode1.head(5)

#PCode1.shape

# ### Substituindo 'Not assigned' por "Queen's Park"

# In[15]:

PCode1.loc[PCode1['Neighbourhood']=='Not assigned']

PCode1.loc[85,'Neighbourhood']= "Queen's Park"

PCode1.loc[PCode1['Postcode']=='M7A']

#PCode1.head(5)

PCode1.shape

GeoCan=pd.read_csv('http://cocl.us/Geospatial_data')
GeoCan.head(5)

PCodeGeo=PCode1.merge(GeoCan, left_on='Postcode',right_on='Postal Code')
PCodeGeo.drop('Postal Code', axis=1, inplace=True)
PCodeGeo

### Abaixo estão o agrupamento de todos os bairros com o nome "Toronto".

# PCodeGeo.loc[(PCodeGeo['Borough']=='East Toronto') | (PCodeGeo['Borough']=='Downtown Toronto') | (PCodeGeo['Borough']=='West Toronto')]
```
